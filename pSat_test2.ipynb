```python
from pSAT import Solver, Rössler, RK4, ForwardEuler

mySolver = Solver(Rössler(a=0.5), RK4())
mySolver.solve(time_limit=50)



import numpy as np
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
fig = plt.figure(figsize = (8,8))
ax = plt.axes(projection='3d')
ax.grid()
x = [s[0] for s in mySolver.traj]
y = [s[1] for s in mySolver.traj]
z = [s[2] for s in mySolver.traj]

ax.plot3D(x, y, z)
ax.set_title('3D Parametric Plot')

# Set axes label
ax.set_xlabel('x', labelpad=20)
ax.set_ylabel('y', labelpad=20)
ax.set_zlabel('t', labelpad=20)

plt.show()
```

```python
from pSAT import Solver, Lorenz, RK4, ForwardEuler
import numpy as np

mySolver = Solver(Lorenz(), RK4())
mySolver.s = np.array([-12.21263922, -20.67997494, 19.42088604])
mySolver.solve(time_limit=15, upper_h_bound=0.001, save_frequency=5)
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
fig = plt.figure(figsize = (8,8))
ax = plt.axes(projection='3d')
ax.grid()
x = [s[0] for s in mySolver.traj]
y = [s[1] for s in mySolver.traj]
z = [s[2] for s in mySolver.traj]

ax.plot3D(x, y, z)
ax.set_title('A Lorenz rendszer')

# Set axes label
ax.set_xlabel('x', labelpad=20)
ax.set_ylabel('y', labelpad=20)
ax.set_zlabel('t', labelpad=20)
plt.show()
```

```python
from pSAT import Solver, Lorenz, RK4, ForwardEuler
import numpy as np

mySolver2 = Solver(Lorenz(), RK4())
mySolver2.s = np.array([-12.21263922, -20.67997494, 19.42088604])
mySolver2.solve(time_limit=0.00005, upper_h_bound=0.0001, save_frequency=2)
```

```python
import matplotlib.pyplot as plt
r1 = [r[0] for r in mySolver2.axis_lengths]
r2 = [r[1] for r in mySolver2.axis_lengths]
r3 = [r[2] for r in mySolver2.axis_lengths]
t = mySolver2.times
plt.plot(t, r1, label='1')
plt.plot(t, r2, label='2')
plt.plot(t, r3, label='3')
plt.legend()
plt.grid(True)
plt.show()
```
