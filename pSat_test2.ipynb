```python
from pSAT import Solver, Rössler, RK4, ForwardEuler

mySolver = Solver(Rössler(a=0.5), RK4())
mySolver.solve(time_limit=500, tangental_flag=False)



import numpy as np
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
fig = plt.figure(figsize = (8,8))
ax = plt.axes(projection='3d')
ax.grid()
x = [s[0] for s in mySolver.traj]
y = [s[1] for s in mySolver.traj]
z = [s[2] for s in mySolver.traj]

ax.plot3D(x, y, z)
ax.set_title('3D Parametric Plot')

# Set axes label
ax.set_xlabel('x', labelpad=20)
ax.set_ylabel('y', labelpad=20)
ax.set_zlabel('t', labelpad=20)

plt.show()
```

```python
from pSAT import Solver, Rössler, RK4, ForwardEuler

mySolver = Solver(Rössler(a=0.5), RK4())
result = mySolver.get_Lyapunov_convergence(max_time = 50)

import matplotlib.pyplot as plt
plt.grid(True)
plt.title('Omega_ii over time')
t = [x[1] for x in result]
l1 = [sorted(x[0])[2] for x in result]
l2 = [sorted(x[0])[1] for x in result]
l3 = [sorted(x[0])[0] for x in result]
plt.plot(t, l1, label= 'l'+str(1))
plt.plot(t, l2, label= 'l'+str(2))
plt.plot(t, l3, label= 'l'+str(3))
plt.legend()
plt.show()
```

```python
from pSAT import Solver, Lorenz, RK4, ForwardEuler
import numpy as np

mySolver = Solver(Lorenz(), RK4())
mySolver.s = np.array([0.5, 0.5, 0.5])
mySolver.solve(time_limit=50, tangental_flag=False, upper_h_bound=0.001)

from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
fig = plt.figure(figsize = (8,8))
ax = plt.axes(projection='3d')
ax.grid()
x = [s[0] for s in mySolver.traj]
y = [s[1] for s in mySolver.traj]
z = [s[2] for s in mySolver.traj]

ax.plot3D(x, y, z)
ax.set_title('A Lorenz rendszer')

# Set axes label
ax.set_xlabel('x', labelpad=20)
ax.set_ylabel('y', labelpad=20)
ax.set_zlabel('t', labelpad=20)

plt.show()
```
